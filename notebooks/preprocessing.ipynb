```python
# I collected and cleaned each country's dataset manually

# Here's how I merged different country datasets into one master CSV file

import pandas as pd
import os

# Directory where your Excel files are stored
file_directory = r"C:\Users\ASUS\Documents\Projects\World economy data"

# List all Excel file paths in the directory

file_order = [
    "US table.xlsx", "Germany table.xlsx", "China table.xlsx", "Japan table.xlsx",
    "Brazil table.xlsx", "India table.xlsx", "South Africa table.xlsx", "UK table.xlsx",
    "France table.xlsx", "Australia table.xlsx"
]
excel_files = [os.path.join(file_directory, file) for file in file_order]


# Initialize an empty list to store DataFrames
dataframes = []

# Loop through each Excel file and load the data
for file in excel_files:
    # Load the Excel file into a DataFrame
    df = pd.read_excel(file)
    
    # Append the DataFrame to the list
    dataframes.append(df)

# Concatenate all DataFrames
master_df = pd.concat(dataframes, ignore_index=True)



# Sort the DataFrame by 'Year' first and then by 'Country'
master_df.sort_values(by=["Year", "Country"], inplace=True)

# Reset the index for clarity
master_df.reset_index(drop=True, inplace=True)



# Check the structure of the merged DataFrame
print(master_df.info())
print(master_df.head())

# Display rows 34 to 44
print(master_df.iloc[4:11])

# Save the consolidated DataFrame to a CSV (optional)
master_df.to_csv("master_data.csv", index=False)
```
